# Question 1 What is Simple Linear Regression (SLR)? Explain its purpose.


Simple Linear Regression (SLR) is a statistical method used to model the relationship between two continuous variables. It aims to find a linear equation that best describes how a dependent variable (Y) changes in response to an independent variable (X).

Purpose of SLR:

 Prediction: To predict the value of the dependent variable based on the value of the independent variable.

Understanding Relationships: To understand the strength and direction of the linear relationship between two variables.

Quantifying Impact: To quantify how much the dependent variable is expected to change for a unit increase in the independent variable.

# Question 2: What are the key assumptions of Simple Linear Regression?

The key assumptions of Simple Linear Regression (SLR) are:

1. Linearity: The relationship between the independent variable (X) and the dependent variable (Y) is linear. This means that the data points should generally follow a straight line.

2.Independence of Errors: The errors (residuals) are independent of each other. This means that the error for one observation does not influence the error for another observation.

3.Homoscedasticity: The variance of the errors is constant across all levels of the independent variable. In other words, the spread of the residuals should be roughly the same for all predicted values.

4.Normality of Errors: The errors (residuals) are normally distributed. This assumption is particularly important for constructing confidence intervals and performing hypothesis tests.

5.No Multicollinearity (for Multiple Linear Regression, but implicitly relevant): While strictly for Multiple Linear Regression, for SLR, it implies that the independent variable should not be perfectly correlated with itself (which is trivial but ensures it's a valid variable).


# Question 3: Write the mathematical equation for a simple linear regression model and explain each term.

To fit a Simple Linear Regression model in Python, we'll use the `scikit-learn` library, which is a popular machine learning library. We will:

1.  **Generate Sample Data**: Create some synthetic data points for our independent variable (X) and dependent variable (Y) that have a linear relationship.
2.  **Import `LinearRegression`**: Bring in the `LinearRegression` model from `sklearn.linear_model`.
3.  **Reshape Data**: `scikit-learn` models typically expect input features (X) to be 2D arrays, even if there's only one feature. We'll reshape our X data accordingly.
4.  **Initialize and Fit the Model**: Create an instance of `LinearRegression` and then use its `fit()` method to train the model on our data.
5.  **Extract Coefficients**: After fitting, we can access the `intercept_` and `coef_` attributes to see the estimated parameters ($\beta_0$ and $\beta_1$) of our linear model.

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# 1. Generate Sample Data
np.random.seed(0) # for reproducibility
X = np.random.rand(100, 1) * 10 # 100 values between 0 and 10
Y = 2 * X + 5 + np.random.randn(100, 1) * 2 # Y = 2X + 5 + some random noise

# 2. Initialize and Fit the Model
model = LinearRegression()
model.fit(X, Y)

# 3. Extract Coefficients
intercept = model.intercept_[0]
coefficient = model.coef_[0][0]

print(f"Fitted Linear Regression Model:")
print(f"Intercept (beta_0): {intercept:.2f}")
print(f"Coefficient (beta_1): {coefficient:.2f}")

# 4. Visualize the results
plt.figure(figsize=(10, 6))
plt.scatter(X, Y, s=20, alpha=0.7, label='Original Data')
plt.plot(X, model.predict(X), color='red', linewidth=2, label=f'Fitted Line: Y = {coefficient:.2f}X + {intercept:.2f}')
plt.xlabel('Independent Variable (X)')
plt.ylabel('Dependent Variable (Y)')
plt.title('Simple Linear Regression Fit')
plt.legend()
plt.grid(True)
plt.show()
```

# Question 5: What is the method of least squares in linear regression?


The Method of Least Squares is a standard approach in linear regression to estimate the coefficients (like $\beta_0$$\beta_0$ and $\beta_1$$\beta_1$ in simple linear regression) that best fit the data. The core idea is to find the line that minimizes the sum of the squared differences between the observed values and the values predicted by the linear model.

Here's a breakdown:

Residuals (Errors): For each data point, there's a difference between its actual Y value and the Y value predicted by our regression line ($\hat{Y}$$\hat{Y}$). This difference is called a residual or error ($e = Y - \hat{Y}$$e = Y - \hat{Y}$). Some residuals will be positive (actual value is above the line), and some will be negative (actual value is below the line).

Squaring the Residuals: To prevent positive and negative errors from canceling each other out, and to penalize larger errors more heavily, we square each residual.

Sum of Squared Residuals (SSR): We then sum up all these squared residuals for every data point. This sum is often denoted as SSR or RSS (Residual Sum of Squares).

$SSR = \sum_{i=1}^{n} (Y_i - \hat{Y}_i)^2$

Where:

$Y_i$ is the actual observed value for the i-th data point.
$\hat{Y}_i$ is the predicted value for the i-th data point, given by the regression equation (e.g., $\hat{Y}_i = \beta_0 + \beta_1X_i$).
Minimization: The objective of the Method of Least Squares is to find the values of the coefficients ($\beta_0$$\beta_0$ and $\beta_1$$\beta_1$) that make this Sum of Squared Residuals (SSR) as small as possible. In other words, it seeks the line that comes closest to all the data points by minimizing the total squared vertical distances from the points to the line.

By minimizing the sum of squared errors, the least squares method provides a statistically optimal way to determine the regression line's parameters under certain assumptions.



# Question 6: What is Logistic Regression? How does it differ from Linear Regression?

Logistic Regression is a statistical model primarily used for binary classification tasks. Unlike Simple Linear Regression which predicts a continuous outcome, Logistic Regression predicts the probability that an instance belongs to a certain class (e.g., 0 or 1, true or false, yes or no).

Here's how it generally works:

It uses a sigmoid (logistic) function to map any real-valued number into a probability between 0 and 1.
It models the relationship between the independent variables and the log-odds of the dependent variable.
How it differs from Linear Regression:
Type of Dependent Variable:

Linear Regression: Predicts a continuous dependent variable (e.g., house price, temperature).
Logistic Regression: Predicts a categorical (usually binary) dependent variable, representing the probability of a specific outcome (e.g., spam/not spam, disease/no disease).
Equation Output:

Linear Regression: The output is a direct value in the range $(-\infty, \infty)$ that directly predicts the dependent variable ($Y = \beta_0 + \beta_1X$).
Logistic Regression: The output is a probability between 0 and 1, obtained by applying a sigmoid function to a linear combination of predictors ($P(Y=1|X) = \frac{1}{1 + e^{-(\beta_0 + \beta_1X)}} $).
Underlying Function:

Linear Regression: Uses a simple linear function.
Logistic Regression: Uses a sigmoid (logistic) activation function to transform the linear output into a probability.
Error Distribution/Assumptions:

Linear Regression: Assumes normally distributed errors (residuals).
Logistic Regression: Does not assume normally distributed errors; it typically assumes a binomial distribution for the errors.
Estimation Method:

Linear Regression: Uses Ordinary Least Squares (OLS) to minimize the sum of squared errors.
Logistic Regression: Uses Maximum Likelihood Estimation (MLE) to maximize the likelihood of observing the actual data.


# Question 7: Name and briefly describe three common evaluation metrics for regression models.


Certainly! Here are three common evaluation metrics for regression models:

Mean Absolute Error (MAE):

Description: MAE measures the average magnitude of the errors in a set of predictions, without considering their direction. It's the average of the absolute differences between predictions and actual observations. MAE is less sensitive to outliers than MSE.
Formula: $MAE = \frac{1}{n} \sum_{i=1}^{n} |Y_i - \hat{Y}_i|$
Mean Squared Error (MSE):

Description: MSE measures the average of the squares of the errors. It penalizes larger errors more heavily than MAE due to the squaring. It is one of the most common metrics and is useful when you want to heavily penalize large errors.
Formula: $MSE = \frac{1}{n} \sum_{i=1}^{n} (Y_i - \hat{Y}_i)^2$
Root Mean Squared Error (RMSE):

Description: RMSE is the square root of the MSE. It's often preferred over MSE because it has the same units as the dependent variable, making it easier to interpret. Like MSE, it gives higher weight to larger errors.
Formula: $RMSE = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (Y_i - \hat{Y}_i)^2}$
R-squared ($R^2$$R^2$):

Description: R-squared, also known as the coefficient of determination, represents the proportion of the variance in the dependent variable that is predictable from the independent variable(s). It ranges from 0 to 1, where 1 indicates that the model explains all the variability of the response data around its mean.
Formula: $R^2 = 1 - \frac{\sum_{i=1}^{n} (Y_i - \hat{Y}i)^2}{\sum{i=1}^{n} (Y_i - \bar{Y})^2}$ (where $\bar{Y}$ is the mean of the observed Y values)


# Question 8: What is the purpose of the R-squared metric in regression analysis?

R-squared ($R^2$$R^2$), also known as the coefficient of determination, is a key metric in regression analysis that serves the following purposes:

Explaining Variance: It represents the proportion of the variance in the dependent variable (Y) that can be explained by the independent variable(s) (X) in the regression model. In simpler terms, it tells you how well your independent variables explain the variability of your dependent variable.

Goodness of Fit: It is often interpreted as a measure of the goodness of fit of the model. A higher R-squared value indicates that the model fits the data better.

Range: R-squared values typically range from 0 to 1 (or 0% to 100%).

An $R^2$ of 0 indicates that the model explains none of the variability of the dependent variable around its mean.
An $R^2$ of 1 (or 100%) indicates that the model explains all the variability of the dependent variable, meaning the model perfectly fits the data.
Model Comparison (with caution): It can be used to compare different regression models on the same dataset, though it's important to be cautious as adding more independent variables will often increase $R^2$$R^2$ even if those variables are not truly useful (this is where Adjusted R-squared becomes more relevant for multiple regression).

# Question 10: How do you interpret the coefficients in a simple linear regression model?


In a simple linear regression model, represented by the equation $Y = \beta_0 + \beta_1X + \epsilon$$Y = \beta_0 + \beta_1X + \epsilon$:

Intercept ($\beta_0$$\beta_0$): This represents the predicted value of the dependent variable (Y) when the independent variable (X) is zero. In some contexts, this might have a meaningful interpretation (e.g., base salary with zero years of experience), while in others, it might not be practically relevant or within the observed range of X.

Coefficient ($\beta_1$$\beta_1$): This is the slope of the regression line. It tells you the expected change in the dependent variable (Y) for every one-unit increase in the independent variable (X), assuming all other factors remain constant. A positive coefficient indicates a positive relationship (as X increases, Y increases), while a negative coefficient indicates a negative relationship (as X increases, Y decreases).
